### Write a sql query to add a new category in categories table and add 5 books to this category in books table with title, rate and price

```python
conection = sqlite3.connect('Books.db')
sql = 'insert into Gategories (GategoryName) values (?)'
Data = ['Technology']
rows1 = conection.execute(sql,Data)
conection.commit()
conection.close()

q = rows1.lastrowid

Books = [["Delete",18.69,5,q],
        ["Deviced!: Balancing Life and Technology in a Digital World ",30, 4,q],
        ["The Truth About Your Future", 18.9, 4,q],
        ["Digital Leadership: Changing Paradigms for Changing Times", 21.71, 4,q],
        ["How to Be Good at Science, Technology, and Engineering ",42.73, 3,q]]

for Book in Books:

    conection = sqlite3.connect('Books.db')        
    sql = 'insert into Books (Title,Price,Rate,GategoryId) values (?, ?, ?, ?)'
    Data = Book
    rows = conection.execute(sql,Data)
    conection.commit()
    conection.close()
```

### Write a sql query to get the minimum & maximum price for all the books that have rate 5

```python
conection = sqlite3.connect('Books.db')
sql = '''select * from Books where Rate = ? and Price = (select min(Price) from Books where Rate = ?) union
         select * from Books where Rate = ? and Price = (select max(Price) from Books where Rate = ?)'''
Data = [5,5,5,5]
rows = conection.execute(sql,Data).fetchall()
conection.close()
rows
```

### Write a sql query to calculate how many books each category has having count > 10

```python
conection = sqlite3.connect('Books.db')
sql = ''' select Books.GategoryId, Gategories.GategoryName, count(Books.GategoryId) from Books inner join Gategories
          on Books.gategoryId = Gategories.gategoryId 
          group by Books.GategoryId having count(Books.GategoryId) > ? '''
Data = [10]
rows = conection.execute(sql,Data).fetchall()
conection.close()
rows
```

###  Write a sql query to get all books with category_name='Music' using subquery

```python
conection = sqlite3.connect('Books.db')
sql = ''' select Books.Title, Books.Price, Books.Rate, Gategories.GategoryName from Books inner join Gategories
          on Books.gategoryId = Gategories.gategoryId
          where Gategories.GategoryId in (select Gategories.GategoryId from Gategories where GategoryName like ?)'''
Data = ['%Music%']
rows = conection.execute(sql,Data).fetchall()
conection.close()
rows
```
